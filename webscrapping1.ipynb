Crawling

```python
root = 'https://letterboxd.com/film/longlegs/reviews/page/'

usernames = []
reviews = []
dates = []

for page in range(1, 10):
    url = requests.get(root + str(page))
    soup = BeautifulSoup(url.text, 'html.parser')

    for u in soup.find_all('a', {'class':'avatar -a40'}):
        usernames.append(u['href'].strip('/').replace('/', ''))

    for r in soup.find_all('div', {'class':'body-text -prose collapsible-text'}):
        reviews.append(r.find('p').text)
    
    for d in soup.find_all('span', {'class': '_nobr'}):
        dates.append(d.text.strip())
    
```

```python
len(usernames), len(reviews), len(dates)
```

```python
import pandas as pd
df = pd.DataFrame(
    {'Usernames':usernames,
     'Reviews':reviews,
     'Date':dates
     })
df.head()
```
